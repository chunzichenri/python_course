```python
c = []
m = []
g = 9.81
v0 = 15
theta = 60
x = 0
dx = 0.1
y0 = 1
while x <= 1:
    c.append(x)
    x = x + dx
print(c)    
from math import pi, tan, cos
v0 = v0/3.6
theta = theta*pi/180
for x in c:
    y = x*tan(theta) - 1/(2*v0**2)*g*x**2/((cos(theta))**2) + y0
    m.append(y)
print(m)
```

```python
import matplotlib.pyplot as plt
plt.plot(c, m, 'rh-')
```
